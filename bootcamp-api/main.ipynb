# Crear la carpeta del proyecto

### 1. Abre Visual Studio Code
### 2. Click en:
    File -> Open Folder
### 3. Crea una carpeta llamada:
    Bootcamp.api
    Nuestro proyecto empezara asi
### 4. Crear los arcivos de el proyecto:
    Dentro de la carpeta crea:

    bootcamp-api
    │
    ├── main.py
    └── sales.csv
