# Bootcamp API – Verificación del Dataset (Pasos 1–4)

Este notebook sirve para **comprobar que el dataset funciona correctamente antes de usarlo en la API con FastAPI**.

Vamos a comprobar:
1. Que las columnas existen.
2. Que la columna `precio` es numérica.
3. Que el CSV se carga correctamente.
4. Que podemos calcular las ventas totales.


## 1️⃣ Cargar librerías y leer el dataset

Primero cargamos **pandas** y leemos el archivo `ventas.csv`.

````python
import pandas as pd

df = pd.read_csv('ventas.csv')
df.head()
````

## 2️⃣ Comprobar que las columnas existen

Esto nos permite verificar que la columna `precio` existe realmente.

Si aparece algo como `precio ` (con espacio) puede causar errores en la API.

````python
print(df.columns)
````

Si ves espacios extra en los nombres de columnas, puedes limpiarlos así:

````python
df.columns = df.columns.str.strip()
print(df.columns)
````

## 3️⃣ Comprobar el tipo de datos

Aquí verificamos si `precio` es numérico.

Si aparece como `object`, pandas lo está leyendo como texto.

````python
df.dtypes
````

Si `precio` no es numérico, podemos convertirlo:

````python
df['precio'] = pd.to_numeric(df['precio'])
df.dtypes
````

## 4️⃣ Calcular ventas totales

Este paso reproduce exactamente lo que hará el endpoint de la API:

`/ventas/total`

````python
total_sales = df['precio'].sum()
total_sales
````

Si todo funciona correctamente, este valor será el que devolverá tu endpoint:

```python
@app.get('/ventas/total')
def total_sales():
    return {'total_ventas': df['precio'].sum()}
```
