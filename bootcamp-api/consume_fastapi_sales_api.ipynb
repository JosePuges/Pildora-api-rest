# Consumir la API de Ventas

Este notebook muestra cómo **consumir la API que creaste con FastAPI**.

Endpoints usados:
- `/ventas`
- `/ventas/total`
- `/ventas/por-ciudad`
- `/ventas/top-productos`

Convertiremos las respuestas JSON en **DataFrames para análisis**.

## 1. Importar librerías

```python
import requests
import pandas as pd
```

## 2. Definir la URL base de la API

Si ejecutas la API en local con `uvicorn main:app --reload`, la URL será:

```python
base_url = "http://127.0.0.1:8000"
```

## 3. Obtener todas las ventas (`/sales`)

```python
response = requests.get(base_url + "/ventas")
data = response.json()

df_sales = pd.DataFrame(data)
df_sales.head()
```

Ahora tenemos los datos de la API dentro de un **DataFrame**, listo para análisis.

## 4. Obtener ventas totales (`/sales/total`)

```python
response = requests.get(base_url + "/ventas/total")
total_sales = response.json()

total_sales
```

## 5. Ventas por país (`/sales/by-country`)

```python
response = requests.get(base_url + "/ventas/por-pais")
data_pais = response.json()

df_country = pd.DataFrame(data_pais)
df_country
```

## 6. Productos más vendidos (`/sales/top-products`)

```python
response = requests.get(base_url + "/ventas/top-productos")
top_productos = response.json()

top_productos
```

## 7. Ejemplo de análisis rápido

Podemos usar los datos directamente para análisis.

```python
df_sales.describe()
```

Este flujo demuestra cómo un analista puede:

Dataset → API → Request → DataFrame → Análisis
